```python
## Problem 2 

from scipy.integrate import odeint
import numpy as np 
import matplotlib.pyplot as plt

# function that returns dz/dt
def model(z,t):
    s = z[0]
    h = z[1]
    dsdt = 0.00012*s*h - 0.1*s - 0.05*s
    dhdt = -0.00035*s*h + 0.04*h + 500
    dzdt = [dsdt,dhdt] 
    return dzdt

# initial condition 
z0 = [1000,3000] 

# number of time points 
n = 100

# time points 
t = np.linspace(0,32,n)


# store solution 
s = np.empty_like(t)
h = np.empty_like(t)
# record initial conditions 
s[0] = z0[0]
h[0] = z0[1]

# solve ODE 
for i in range(1,n):
    # spant for next time step 
    tspan = [t[i-1],t[i]]
    # solve for next step 
    z = odeint(model,z0,tspan)
    # store solution for plotting 
    s[i] = z[1][0]
    h[i] = z[1][1]
    # next initial condition 
    z0 = z[1]

# plot results 
 
plt.plot(t,s,'b-',label='s(t)') 
plt.plot(t,h,'r--',label='h(t)')
plt.ylabel('population')
plt.xlabel('weeks from January 2023 to start of September 2023')
plt.title('Population of Salmon and Herring from January 2023 to September 2023')
plt.grid()
plt.legend(loc='best')
plt.show()

## Printing results 
pop_s = s[-1] #at time t=32  # -1 because we count backwards in the array to get the last element
pop_h = h[-1]
print("Population of s at t=32 is:", pop_s)
print("Population of h at t=32 is:", pop_h)

print(h)
print(s)


## using the above initial values we define a new model starting from september
```

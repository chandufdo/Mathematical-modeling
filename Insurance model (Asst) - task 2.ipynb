```python
import numpy as np
import matplotlib.pyplot as plt
import sympy
import math

#Expected value of the Terminal Wealth without Insurance

w = 0.5 #Initial Wealth
g = [1,0] #Gains
p = [2/3,1/3] #Probabilities
N = 10 #Number of farmers
n = 2
gamma = [1,1.5,2,2.5,3,3.5,4,4.5,5,5.5]
E = np.zeros(n)   # expected value of C
C = np.zeros(n)   # C (cover)
W = np.zeros(n)   # Terminal Wealth
U = np.zeros(n)   # Utility value
EU = np.zeros(N)  # Expected Utility value
premium= np.zeros(N)

print("Expected Utility value of each farmer - Without Insurance")

for j in range(N):
    
    for i in range(n):
        
        W[i] = w + g[i]
        U[i] = 1 - math.exp(-gamma[j]*W[i])
        E[i] = U[i] * p[i]

    EU[j] = sum(E)
    print("Expected Utility value of farmer",j+1,"= ", EU[j])

plt.figure(1)
plt.plot(gamma,EU,'*-b',linewidth=2,label='E[U] Without Insurance')
plt.xlabel("Gamma Values")
plt.ylabel("Expected Utility")
plt.legend()
plt.title("Expected Utility Vs Gamma Values - Without Insurance")
plt.show()


#Expected value of the Terminal Wealth with Traditional Insurance

#Part 1 - Full Cover Insurance
print("  ")
print("Premium of Farmers with Full cover Traditional Insurance ")

for j in range(N):
    premium[j] = 1.5 + (math.log((1-EU[j]))/(gamma[j]))
    print("Premium of farmer ",j+1,"= ", premium[j])

print(' ')
print("Expected Utility(Full Cover) Value of each Farmer")
W1 = np.zeros(n)
U1 = np.zeros(n)
EU1 = np.zeros(N)
E1 = np.zeros(n)

for i in range(n):
    if g[i] == 0:
        C[i] = 1
    else:
        C[i] = 0

for j in range(N):
    for i in range(n):
        W1[i] = w - premium[j] + g[i] + C[i]
        U1[i] = 1 - math.exp(-gamma[j]*W1[i])
        E1[i] =U1[i] * p[i]
        #print(E1)

    EU1[j] = sum(E1)

for i in range(N):
    print("Expected value of farmer", i+1, " = ",EU1[i])


plt.figure(2)
plt.plot(gamma,EU1,'*-r',linewidth=2,label='E[U]Trad-insurance FC')
plt.xlabel("Gamma Values")
plt.ylabel("Expected Utility")
plt.legend()
plt.title("Expected Utility Vs Gamma Values - Traditional Insurance(Full Cover)")
plt.show()


#Part 2 - Partial Cover Insurance

print(" ")
print(" Traditional Insurance - Partial Cover")

EU2 = np.zeros(N)
a = sympy.symbols('a')
premium2 = a/3

        
EUP = np.zeros(N)

for j in range(N):
    W21 = float(w) - premium2 + float(g[0]) + 0
    W22 = float(w) - premium2 + float(g[1]) + a
    EU2 = ((1 - sympy.exp(-gamma[j]*W21))*p[0]) + ((1 - sympy.exp(-gamma[j]*W22))*p[1])
    df = sympy.diff(EU2,a)
    solv = sympy.solve(df)
     
    print("Premium paid by farmer ",j+1, solv[0]," ")

#By using the printed values,
c = [1.00000000000000 , 0.999999999999999, 0.999999999999998 , 0.999999999999999 ,  0.999999999999999 , 0.999999999999999 , 1.00000000000000 , 1.00000000000000 , 1.00000000000000 , 1.00000000000000]
#print(c)  # c = a 

for i in range(N):
    W21s = float(w) - (c[i]/3) + float(g[0]) + 0
    W22s = float(w) - (c[i]/3) + float(g[1]) + c[i]
    EUP[i] = ((1 - sympy.exp(-gamma[i]*W21s))*p[0]) + ((1 - sympy.exp(-gamma[i]*W22s))*p[1])
    

for i in range(N):
    print("Expected value of farmer", i+1, " = ",EUP[i])

    
plt.figure(3)
plt.plot(gamma,EUP,'*-g',linewidth=2,label='E[U]Trad-insurance PC')
plt.xlabel("Gamma Values")
plt.ylabel("Expected Utility")
plt.legend()
plt.title("Expected Utility Vs Gamma Values - Traditional Insurance(Partial Cover)")
plt.show()
print(" ")



# Index Insurance (Full Cover)



print(" ")
print("Index Insurance - Full Cover")

n1 = 4    #no. of index probabilities
W3 = np.zeros(n1)
U3 = np.zeros(n1)
EU3 = np.zeros(N)
E3 = np.zeros(n1)

premium3 = 1/3   # Computed mannually from premium = Expected val. of Cover

W3[0] = 2.5 - premium3  # Index bad, year good
W3[1] = 1.5 - premium3  # Index bad, year bad
W3[2] = 1.5 - premium3  # Index good, year good
W3[3]= 0.5 - premium3   # Index good, year bad

p1 = [2/9,1/9,1/9,5/9]  # The 4 index probabilities of [(Ib,Yg),(Ib,Yb),(Ig,Yg),(Ig,Yb)]

for j in range(N):
    for i in range(n1):
        U3[i] = 1 - math.exp(-gamma[j]*W3[i])
        E3[i] = U3[i] * p1[i]

    EU3[j] = sum(E3)
    print("Expected Utility(Full cover of Index Insurace) value of farmer  ",j+1, EU3[j])
    
plt.figure(4)
plt.plot(gamma,EU3,'*-c',linewidth=2,label='E[U]Index-Insurance FC')
plt.xlabel("Gamma Values")
plt.ylabel("Expected Utility")
plt.legend()
plt.title("Expected Utility Vs Gamma Values - Index Insurance(Full Cover)")
plt.show()
print(" ") 




# Index Insurance(partial cover)



print(" ")
print("Index Insurance - Partial Cover")

CoInsLvl = np.zeros(N)   # Coinsurance level array
EU4 = np.zeros(N)        # Expected value in terms of b
EUIP = np.zeros(N)   # Solved Expected value of Index Insurance with partial cover

b = sympy.symbols('b')
premium4 = b/3   # By premium = E(C)
p1 = [2/9,1/9,1/9,5/9]  # The 4 index probabilities of [(Ib,Yg),(Ib,Yb),(Ig,Yg),(Ig,Yb)]


for j in range(N):
    W41 = float(w) - premium4 + float(g[0]) + b  #Terminal Wealth of Ib,Yg
    W42 = float(w) - premium4 + float(g[1]) + b  #Terminal Wealth of Ib,Yb
    W43 = float(w) - premium4 + float(g[0])      #Terminal Wealth of Ig,Yg
    W44 = float(w) - premium4 + float(g[1])      #Terminal Wealth of Ig,Yb
    EU4 = ((1 - sympy.exp(-gamma[j]*W41))*p1[0]) + ((1 - sympy.exp(-gamma[j]*W42))*p1[1]) + ((1 - sympy.exp(-gamma[j]*W43))*p1[2]) + ((1 - sympy.exp(-gamma[j]*W44))*p1[3])
    df1 = sympy.diff(EU4,b)
    solv1 = sympy.solve(df1)
    CoInsLvl[j] = solv1[0]

    #print(j+1, solv1)
    #print(j+1, InsCover[j])
    
for i in range(N):
    print("Premium paid by farmer", i+1, " = ",CoInsLvl[i])

print(" ")
    
for i in range(N):
    W41s = float(w) - (CoInsLvl[i]/3) + float(g[0]) + CoInsLvl[i]    #Terminal Wealth of Ib,Yg
    W42s = float(w) - (CoInsLvl[i]/3) + float(g[1]) + CoInsLvl[i]    #Terminal Wealth of Ib,Yb
    W43s = float(w) - (CoInsLvl[i]/3) + float(g[0])                  #Terminal Wealth of Ig,Yg
    W44s = float(w) - (CoInsLvl[i]/3) + float(g[1])                  #Terminal Wealth of Ig,Yb
    EUIP[i] = ((1 - sympy.exp(-gamma[i]*W41s))*p1[0]) + ((1 - sympy.exp(-gamma[i]*W42s))*p1[1]) + ((1 - sympy.exp(-gamma[i]*W43s))*p1[2]) + ((1 - sympy.exp(-gamma[j]*W44s))*p1[3])
    

for i in range(N):
    print("Expected value of farmer", i+1, " = ",EUIP[i])

    
plt.figure(5)
plt.plot(gamma,EUIP,'*-m',linewidth=2,label='E[U]Index-Insurance PC')
plt.xlabel("Gamma Values")
plt.ylabel("Expected Utility")
plt.legend()
plt.title("Expected Utility Vs Gamma Values - Index Insurance(Partial Cover)")
plt.show()
print(" ")    



# FINAL PLOTS



print("  ")
print("To find which type insurance is suitable for each farmer")

plt.figure(6)
plt.plot(gamma,EU,'*',linewidth=2,label='E[U] Without Insurance')    
plt.plot(gamma,EU1,'-r',linewidth=2,label='E[U]Trad-insurance FC')
plt.plot(gamma,EUP,'*-g',linewidth=2,label='E[U]Trad-insurance PC')
plt.plot(gamma,EU3,'*-c',linewidth=2,label='E[U]Index-Insurance FC')
plt.plot(gamma,EUIP,'*-m',linewidth=2,label='E[U]Index-Insurance PC')
plt.xlabel("Gamma Values")
plt.ylabel("Expected Utility")
plt.legend()
plt.title("Expected Utility Vs Gamma Values - All 4 types")
plt.grid()
plt.show()
print(" ")    
    
    
    
```
